```python
import json

with open('dataset/advglue.json') as f:
    data = json.load(f)

result = []

for key, value in data.items():
    if key == 'sst2':
        for item in value:
            result.append({
                'text': item['sentence'],
                'meta': {
                    'label': item['label'],
                    'source': 'sst2'
                }
            })
    elif key == 'qqp':
        for item in value:
            result.append({
                'text': item['question1'] + '٩' + item['question2'],
                'meta': {
                    'question1': item['question1'],
                    'question2': item['question2'],
                    'label': item['label'],
                    'delimiter': '٩',
                    'source': 'qqp'
                }
            })
    elif key == 'mnli':
        for item in value:
            result.append({
                'text': item['premise'] + '٩' + item['hypothesis'],
                'meta': {
                    'premise': item['premise'],
                    'hypothesis': item['hypothesis'],
                    'label': item['label'],
                    'delimiter': '٩',
                    'source': 'mnli'
                }
            })
    elif key == 'mnli-mm':
        for item in value:
            result.append({
                'text': item['premise'] + '٩' + item['hypothesis'],
                'meta': {
                    'premise': item['premise'],
                    'hypothesis': item['hypothesis'],
                    'label': item['label'],
                    'delimiter': '٩',
                    'source': 'mnli-mm'
                }
            })
    elif key == 'qnli':
        for item in value:
            result.append({
                'text': item['question'] + '٩' + item['sentence'],
                'meta': {
                    'question': item['question'],
                    'sentence': item['sentence'],
                    'label': item['label'],
                    'delimiter': '٩',
                    'source': 'qnli'
                }
            })
    elif key == 'rte':
        for item in value:
            result.append({
                'text': item['sentence1'] + '٩' + item['sentence2'],
                'meta': {
                    'sentence1': item['sentence1'],
                    'sentence2': item['sentence2'],
                    'label': item['label'],
                    'delimiter': '٩',
                    'source': 'rte'
                }
            })

with open('dataset/advglue2.json', 'w') as f:
    json.dump(result, f, indent=2)
```
